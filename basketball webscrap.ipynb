```python
from bs4 import *
```

```python
import requests
data = requests.get("https://www.basketball-reference.com/")
```

```python
soup = BeautifulSoup(data.text,"html.parser")
```

```python
data = soup.find_all("th",scope="row")
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
name = []
```

```python
for i in data:
    i=i.text
    i=i.split('\xa0')
    i=''.join(i)
    name.append(i)
```

```python
name 
```

```python
name = name[:197]
len(name)
```

```python
franchise = []
data_s = soup.find_all("td")
for i in range(0,len(data_s)+1,4):
    i = data_s[i].text
    franchise.append(i)
```

```python
franchise
```

```python
franchise = franchise[:197]
len(franchise)
```

```python
payroll = []
data_s = soup.find_all("td")
for i in range(1,len(data_s)+1,4):
    i = data_s[i].text
    payroll.append(i)
payroll
```

```python
payroll = payroll[:197]
len(payroll)
```

```python
wins = []
data_s = soup.find_all("td")
for i in range(2,len(data_s)+1,4):
    i = data_s[i].text
    wins.append(i)
wins
```

```python
wins = wins[:197]
len(wins)
```

```python
losses = []
data_s = soup.find_all("td")
for i in range(3,len(data_s),4):
    i = data_s[i].text
    losses.append(i)
losses
len(losses)
```

```python
east_standing = pd.DataFrame({'names':name,'franchise':franchise,'payroll':payroll,'wins':wins,'losses':losses})
```
